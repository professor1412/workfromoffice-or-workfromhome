```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset.drop(["ID","Name"],axis=1,inplace=True)
dataset.head()
```

```python
dataset.shape
```

```python
print(dataset["Occupation"].unique())
```

```python
Occupation_mapping= {
    "Tutor" :0,
    "HR": 1,
    "Engineer": 2,
    "Recruiter": 3,
    "Business": 4,
    "Marketing": 5,
    "Manager": 6
}

dataset["Occupation"] = dataset["Occupation"].replace(Occupation_mapping)
dataset.head()
```

```python
dataset["Gender"] = dataset["Gender"].replace({
    "Female": 0,
    "Male": 1
})
dataset.head(10)
```

```python
print(dataset["RM_productive"].min())
print(dataset["RM_productive"].max())
print(dataset["RM_productive"].nunique())
```

```python
dataset= dataset.replace({"Yes": 1, "No": 0, "Not sure": 2})
dataset.head(10)
```

```python
print(dataset["calmer_stressed"].unique())
```

```python
dataset = dataset.replace({"CALMER": 1, "STRESSED": 0})
```

```python
dataset = dataset.dropna(subset=['Same_ofiice_home_location','RM_job_opportunities','Occupation'])
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.dtypes
```

```python
dataset['Occupation']=dataset['Occupation'].astype(int)
dataset['Same_ofiice_home_location']=dataset['Same_ofiice_home_location'].astype(int)
dataset.dtypes
```

```python
x= dataset.iloc[:, :-1]
y= dataset.iloc[:, -1]
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()

x_train = ms.fit_transform(x_train)
x_test = ms.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Naive Bayes': GaussianNB(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Bagging': BaggingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Extra Trees': ExtraTreeClassifier(),
}
```

```python
for name, model in models.items():
    model.fit(x_train,y_train)
    ypred = model.predict(x_test)
    
    print(f"{name}  with accuracy : {accuracy_score(y_test,ypred)}")
    print("Confusion matrix : ",confusion_matrix(y_test,ypred))
    print("==========================================================")
    
```

```python
# finally selected randomforest model

rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
ypred = rfc.predict(x_test)
accuracy_score(y_test,ypred)
```
